## Paso 1: Importar las bibliotecas y cargar los datos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import MDS

# Cargar los datos del archivo morse.csv
morse_data = pd.read_csv("D:\MultivariateMethods\Data\morse.csv")

# Asegurarnos de que la primera fila no sea interpretada como encabezado
print("Dimensiones originales:", morse_data.shape)
```

```python
morse_data
```

## Paso 2: Descomponer la matriz  en sus componentes  y 

```python
# Convertir la matriz de confusin a un numpy array
M = morse_data.values

# Calcular la parte simtrica A
A = 0.5 * (M + M.T)

# Calcular la parte anti-simtrica B
B = 0.5 * (M - M.T)

# Convertir la matriz A a un DataFrame para visualizacin
A_df = pd.DataFrame(A, index=morse_data.index, columns=morse_data.columns)
print("\nMatriz Simtrica (A):")
print(A_df)
```

## Paso 3: Aplicar Escalamiento Multidimensional No Mtrico (NMDS) en la matriz simtrica 


```python
# Aplicar MDS no mtrico en la matriz simtrica A
mds = MDS(n_components=2, dissimilarity='precomputed', metric=False, random_state=42)
mds_fit = mds.fit_transform(1 - A)  # Usamos 1 - A para convertir tasas de confusin en similitud

# Convertir las coordenadas resultantes a un DataFrame
mds_df = pd.DataFrame(mds_fit, index=morse_data.index, columns=["Dim1", "Dim2"])
print("\nCoordenadas del NMDS:")
print(mds_df)
```

## Paso 4: Graficar la solucin obtenida

```python
# Graficar las coordenadas obtenidas
plt.figure(figsize=(10, 8))
plt.scatter(mds_df["Dim1"], mds_df["Dim2"], color='blue')

# Aadir etiquetas para cada seal de Morse
for i, label in enumerate(mds_df.index):
    plt.text(mds_df["Dim1"][i], mds_df["Dim2"][i], label, fontsize=9)

plt.title("Escalamiento Multidimensional No Mtrico - Seales de Morse")
plt.xlabel("Dimensin 1")
plt.ylabel("Dimensin 2")
plt.grid(True)
plt.show()
```

### Interpretacin de Resultados

1. **Descomposicin de la Matriz**:
   - La matriz original \( M \) se descompuso en una parte simtrica \( A \) y una parte anti-simtrica \( B \). La parte simtrica \( A \) refleja la similitud promedio entre las seales cuando se presentan en ambas direcciones (antes y despus).

2. **Anlisis NMDS**:
   - Al aplicar el escalamiento no mtrico sobre la parte simtrica \( A \), hemos obtenido una representacin en un espacio bidimensional que preserva las relaciones de similitud entre las seales de Morse.
   - La visualizacin muestra cmo algunas seales tienden a agruparse, lo que indica que son ms confusas entre s.

3. **Nmero de Dimensiones**:
   - En este anlisis, utilizamos 2 dimensiones para simplificar la visualizacin. Sin embargo, se recomienda revisar el "stress" del modelo para determinar si se necesitan ms dimensiones para capturar mejor la estructura de los datos. Si el "stress" es alto, se podra intentar con 3 dimensiones.


## Paso 6: Evaluar el "Stress" para decidir el nmero de dimensiones

```python
# Evaluar el "stress" en diferentes dimensiones
stress_values = []
dimensions = [1, 2, 3, 4, 5]

for dim in dimensions:
    mds = MDS(n_components=dim, dissimilarity='precomputed', metric=False, random_state=42)
    mds_fit = mds.fit_transform(1 - A)
    stress_values.append(mds.stress_)
    print(f"Stress en {dim} dimensin(es): {mds.stress_}")

# Graficar Stress vs Dimensin
plt.figure(figsize=(8, 6))
plt.plot(dimensions, stress_values, marker='o')
plt.xlabel("Dimensin")
plt.ylabel("Stress")
plt.title("Stress vs Nmero de Dimensiones")
plt.grid(True)
plt.show()
```

### Interpretacin y Conclusiones

1. **Anlisis del "Stress" por Dimensin**:
   - El "stress" obtenido en **1 dimensin** es muy alto (**50.58**), lo que indica que la representacin en una sola dimensin no es adecuada para capturar la estructura de similitud de las seales Morse.
   - Al aumentar a **2 dimensiones**, el "stress" disminuye significativamente a **26.25**, pero sigue siendo relativamente alto, lo que sugiere que la configuracin en 2 dimensiones an no es ptima.
   - En **3 dimensiones**, el "stress" baja a **16.31**, lo que indica una mejora considerable en la representacin. Sin embargo, este valor an puede considerarse algo elevado para una buena representacin.
   - Con **4 dimensiones**, el "stress" se reduce an ms a **10.92**, lo que indica que estamos capturando mejor la estructura de los datos. Este nivel de "stress" empieza a ser aceptable.
   - Finalmente, en **5 dimensiones**, el "stress" alcanza un valor de **8.38**, lo que sugiere que esta configuracin es capaz de representar de forma mucho ms precisa la estructura de similitud original.

2. **Nmero ptimo de Dimensiones**:
   - Basndonos en los valores de "stress", vemos una mejora significativa al pasar de 1 a 3 dimensiones, y luego una disminucin ms moderada al pasar de 3 a 5 dimensiones.
   - La configuracin en **4 dimensiones** parece ser un buen compromiso entre simplicidad y precisin, ya que el "stress" de **10.92** es lo suficientemente bajo como para proporcionar una representacin adecuada sin aadir demasiada complejidad.
   - Sin embargo, si se busca una precisin an mayor, la configuracin en **5 dimensiones** sera la ideal, con un "stress" de **8.38**, que es bastante bajo y permite una representacin ms precisa de las similitudes entre las seales.

3. **Recomendacin Final**:
   - Para una interpretacin prctica y para facilitar la visualizacin, **4 dimensiones** son suficientes y proporcionan una representacin adecuada con un nivel de "stress" razonable.
