## Paso 1: Importar las bibliotecas y cargar los datos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import MDS

# Cargar los datos del archivo morse.csv
morse_data = pd.read_csv("D:\MultivariateMethods\Data\morse.csv")

# Asegurarnos de que la primera fila no sea interpretada como encabezado
print("Dimensiones originales:", morse_data.shape)
```

```python
morse_data
```

## Paso 2: Descomponer la matriz 𝑀 en sus componentes 𝐴 y 𝐵

```python
# Convertir la matriz de confusión a un numpy array
M = morse_data.values

# Calcular la parte simétrica A
A = 0.5 * (M + M.T)

# Calcular la parte anti-simétrica B
B = 0.5 * (M - M.T)

# Convertir la matriz A a un DataFrame para visualización
A_df = pd.DataFrame(A, index=morse_data.index, columns=morse_data.columns)
print("\nMatriz Simétrica (A):")
print(A_df)
```

## Paso 3: Aplicar Escalamiento Multidimensional No Métrico (NMDS) en la matriz simétrica 𝐴


```python
# Aplicar MDS no métrico en la matriz simétrica A
mds = MDS(n_components=2, dissimilarity='precomputed', metric=False, random_state=42)
mds_fit = mds.fit_transform(1 - A)  # Usamos 1 - A para convertir tasas de confusión en similitud

# Convertir las coordenadas resultantes a un DataFrame
mds_df = pd.DataFrame(mds_fit, index=morse_data.index, columns=["Dim1", "Dim2"])
print("\nCoordenadas del NMDS:")
print(mds_df)
```

## Paso 4: Graficar la solución obtenida

```python
# Graficar las coordenadas obtenidas
plt.figure(figsize=(10, 8))
plt.scatter(mds_df["Dim1"], mds_df["Dim2"], color='blue')

# Añadir etiquetas para cada señal de Morse
for i, label in enumerate(mds_df.index):
    plt.text(mds_df["Dim1"][i], mds_df["Dim2"][i], label, fontsize=9)

plt.title("Escalamiento Multidimensional No Métrico - Señales de Morse")
plt.xlabel("Dimensión 1")
plt.ylabel("Dimensión 2")
plt.grid(True)
plt.show()
```

### Interpretación de Resultados

1. **Descomposición de la Matriz**:
   - La matriz original \( M \) se descompuso en una parte simétrica \( A \) y una parte anti-simétrica \( B \). La parte simétrica \( A \) refleja la similitud promedio entre las señales cuando se presentan en ambas direcciones (antes y después).

2. **Análisis NMDS**:
   - Al aplicar el escalamiento no métrico sobre la parte simétrica \( A \), hemos obtenido una representación en un espacio bidimensional que preserva las relaciones de similitud entre las señales de Morse.
   - La visualización muestra cómo algunas señales tienden a agruparse, lo que indica que son más confusas entre sí.

3. **Número de Dimensiones**:
   - En este análisis, utilizamos 2 dimensiones para simplificar la visualización. Sin embargo, se recomienda revisar el "stress" del modelo para determinar si se necesitan más dimensiones para capturar mejor la estructura de los datos. Si el "stress" es alto, se podría intentar con 3 dimensiones.


## Paso 6: Evaluar el "Stress" para decidir el número de dimensiones

```python
# Evaluar el "stress" en diferentes dimensiones
stress_values = []
dimensions = [1, 2, 3, 4, 5]

for dim in dimensions:
    mds = MDS(n_components=dim, dissimilarity='precomputed', metric=False, random_state=42)
    mds_fit = mds.fit_transform(1 - A)
    stress_values.append(mds.stress_)
    print(f"Stress en {dim} dimensión(es): {mds.stress_}")

# Graficar Stress vs Dimensión
plt.figure(figsize=(8, 6))
plt.plot(dimensions, stress_values, marker='o')
plt.xlabel("Dimensión")
plt.ylabel("Stress")
plt.title("Stress vs Número de Dimensiones")
plt.grid(True)
plt.show()
```

### Interpretación y Conclusiones

1. **Análisis del "Stress" por Dimensión**:
   - El "stress" obtenido en **1 dimensión** es muy alto (**50.58**), lo que indica que la representación en una sola dimensión no es adecuada para capturar la estructura de similitud de las señales Morse.
   - Al aumentar a **2 dimensiones**, el "stress" disminuye significativamente a **26.25**, pero sigue siendo relativamente alto, lo que sugiere que la configuración en 2 dimensiones aún no es óptima.
   - En **3 dimensiones**, el "stress" baja a **16.31**, lo que indica una mejora considerable en la representación. Sin embargo, este valor aún puede considerarse algo elevado para una buena representación.
   - Con **4 dimensiones**, el "stress" se reduce aún más a **10.92**, lo que indica que estamos capturando mejor la estructura de los datos. Este nivel de "stress" empieza a ser aceptable.
   - Finalmente, en **5 dimensiones**, el "stress" alcanza un valor de **8.38**, lo que sugiere que esta configuración es capaz de representar de forma mucho más precisa la estructura de similitud original.

2. **Número Óptimo de Dimensiones**:
   - Basándonos en los valores de "stress", vemos una mejora significativa al pasar de 1 a 3 dimensiones, y luego una disminución más moderada al pasar de 3 a 5 dimensiones.
   - La configuración en **4 dimensiones** parece ser un buen compromiso entre simplicidad y precisión, ya que el "stress" de **10.92** es lo suficientemente bajo como para proporcionar una representación adecuada sin añadir demasiada complejidad.
   - Sin embargo, si se busca una precisión aún mayor, la configuración en **5 dimensiones** sería la ideal, con un "stress" de **8.38**, que es bastante bajo y permite una representación más precisa de las similitudes entre las señales.

3. **Recomendación Final**:
   - Para una interpretación práctica y para facilitar la visualización, **4 dimensiones** son suficientes y proporcionan una representación adecuada con un nivel de "stress" razonable.
   - Si se requiere una representación con la menor distorsión posible y la visualización no es una prioridad, se podría optar por **5 dimensiones** para reducir aún más el "stress".

Por lo tanto, la elección entre 4 y 5 dimensiones dependerá del balance que se quiera lograr entre la simplicidad y la precisión en la representación de los datos.
